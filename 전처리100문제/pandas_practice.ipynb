```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'

df = pd.read_csv(DataUrl)
df.head()
```

```python
# Q45 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라
a = df['host_name'].value_counts()
b = pd.DataFrame(a).sort_values(by='host_name')
print(b)
```

```python
# Q46 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라 
# a = df['host_name'].value_counts()
# ans = pd.DataFrame(a)
# ans

Ans = df.groupby('host_name').size().\
Ans
```

```python
# Q47 neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라
a = df.groupby(['neighbourhood_group','neighbourhood']).size()
a.head()
```

```python
a = df[['neighbourhood_group','neighbourhood']].groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
b = a.groupby('neighbourhood_group', as_index=False).max()
print(a)
print(b)
```

```python
# Q49 neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라
a = df.groupby('neighbourhood_group')['price'].agg(['mean','var', 'max', 'min'])
print(a)
```

```python
# Q50 neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라
a = df.groupby('neighbourhood_group')['reviews_per_month'].agg(['mean','var','max','min'])
print(a)
```

```python
# Q51 neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 구하라
a = df.groupby(['neighbourhood','neighbourhood_group'])['price'].mean()
print(a)
```

```python
# Q52 neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 계층적 indexing 없이 구하라
a = df.groupby(['neighbourhood','neighbourhood_group'])['price'].mean().unstack()
print(a)
```

```python
# Q53 neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라
a = df.groupby(['neighbourhood','neighbourhood_group'])['price'].mean().unstack()
b = a.fillna(-999) 
print(b)
```

```python
# Q54 데이터 중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라
condition = df.neighbourhood_group=='Queens'
a = pd.DataFrame(df[condition])
b = a.groupby('neighbourhood')['price'].agg(['mean','var','max','min'])
print(b)
```

```python
# Q55 데이터 중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라
a = df.groupby(['neighbourhood_group','room_type']).size().unstack()
print(a)
```
